```python
# write a to calculate the avg of 4 number using function
def avgs(a,b,c,d):
    avg=(a+b+c+d)/4
    print(f'The avg of four number is :{avg}')
    return avg

avgs(2,4,5,7)

```

```python
#calculate the area of circle using fucntion
def cir(radius):
    area=3.14*(radius**2)
    print(f'the area of circle is {area}')
    return area
rad=int(input("enter the radius of circle: "))
cir(rad)
```

```python
#to check number if prime or not
def check(num):
    if num>=1:
      for i in range(2, num):
        if num % i == 0:
           print("num is prime")
        else:
           print("num is not prime")
check(9)
```
